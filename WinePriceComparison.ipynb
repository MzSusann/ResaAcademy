```python
# ---------------------------- WINE HOUSE NIGERIA ----------------------- #

#Import tools

import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from time import sleep
from random import randint


pages = np.arange(1, 12, 1) #Because we have 11 pages that we want to iterate on

#initialize empty lists where we'll store our data
images = []
names = []
price = []

for page in pages: 

  page = requests.get("https://winehousenigeria.com/product-category/wine/page/" + str(page) + "/")

  soup = BeautifulSoup(page.text, 'html.parser')
  wine_div = soup.find_all('li', class_= "product")
  
  sleep(randint(2,10))

  for container in wine_div:
      name = container.a.find("h2", class_="woocommerce-loop-product__title").text
      names.append(name)

      image = container.a.find("span", class_="et_shop_image").img["src"]
      images.append(image)

      cost = container.a.find("span", class_="woocommerce-Price-amount").text
      price.append(cost)
    
        
wine_house_df = pd.DataFrame({"Name":names, "Images": images, "Price": price})
wine_house_df.head()

wine_house_df.to_csv("WineHouseNigeria.csv")
```

```python
#------------------- DRINKSNG ----------------------------- #


drinkpages = np.arange(1, 13, 1) #Because we have 12 pages that we want to iterate on

drinknames = [] #store the data we want to extract
drinkimages = []
drinkprices = []

for drinkpage in drinkpages:
    drinkpage = requests.get("https://www.drinks.ng/shop/wine/page/" + str(drinkpage) + "/" )
    drinksoup = BeautifulSoup(drinkpage.text, "html.parser")
    
    drinkwine_div = drinksoup.find_all('div', class_= "prod") #contains the container with the name, images and price
    drinkwine_divv = drinksoup.find_all('div', class_ = "content-exta col-lg-3 col-md-12") #contains the price container only of the wine

    sleep(randint(2,10)) #between 2 to 10 secs
        
    for i1,i2 in zip(drinkwine_div, drinkwine_divv):
        drinkname = i1.div.find('h4', class_ = "woocommerce-loop-product__title title").get_text()
        drinknames.append(drinkname)
        
        drinkimage = i1.div.find('a', class_ = "woocommerce-LoopProduct-link woocommerce-loop-product__link").img["src"]
        drinkimages.append(drinkimage)
        
        drinkprice = i2.div.find('span', class_ = "woocommerce-Price-amount amount").get_text()
        drinkprices.append(drinkprice)
    
drinkdf = pd.DataFrame({"DrinkNG Names: ": drinknames, "DrinkNG Images: ": drinkimages, "DrinkNG Prices: ": drinkprices})
drinkdf.head()
    
drinkdf.to_csv("DrinkNG.csv")
```

```python
# -------------------------- DRINK IN LAGOS ---------------------- #
drinklgnames = []
drinklgimages = []
drinklgprices = []

drinklgurl = requests.get("https://www.drinksinlagos.com/product-category/wines/")
drinklgsoup = BeautifulSoup(drinklgurl.text, 'html.parser')


    
drinklg = drinklgsoup.find_all('div', class_ = "post_item_wrap")

sleep(randint(2,10))

for i in drinklg:
    drinklgname = i.find('h2', class_ = "woocommerce-loop-product__title").text
    drinklgnames.append(drinklgname)
    
    drinklgimage = i.find('div', class_ = "post_thumb").img['src']
    drinklgimages.append(drinklgimage)
    
    drinklgprice = i.find('span', class_ = "woocommerce-Price-amount amount").text
    drinklgprices.append(drinklgprice)

#print(drinklgnames, drinklgimages, drinklgprices)

drinklgdf = pd.DataFrame({"Drinks In Lagos Name ": drinklgnames, "Drinks In Lagos Images ": drinklgimages, "Drinks In Lagos Prices ": drinklgprices,})
drinklgdf.head()

drinklgdf.to_csv("Drinks In Lagos.csv")
```

```python
jumiapages = np.arange(1, 6, 1) #Because we have 5 pages

jumiawinenames = []
jumiawineimages = []
jumiawineprices = []

for jumiapage in jumiapages:
    jumiaurl = requests.get("https://www.jumia.com.ng/wines/?page="+str(jumiapage))
    jumiasoup = BeautifulSoup(jumiaurl.text, "html.parser") #products -mabaya
    
    jumiawine = jumiasoup.find_all('div', class_ = 'sku -gallery')
    
    sleep(randint(2,10)) #between 2 to 10 secs
    
    for jumia in jumiawine:
        jumiawinename = jumia.find('span', class_= 'name').text
        jumiawinenames.append(jumiawinename)
        
        jumiawineimg = jumia.find('div', class_ = 'image-wrapper default-state').img['data-src']
        jumiawineimages.append(jumiawineimg)
        
        jumiawineprice = jumia.find('span', class_ = 'price-box ri').text
        jumiawineprices.append(jumiawineprice)
        

jumiawinedf = pd.DataFrame({"Jumia Wine Names:": jumiawinenames, "Jumia Wine Images:": jumiawineimages, "Jumia Wine Prices:": jumiawineprices})
jumiawinedf.head()
    
jumiawinedf.to_csv("Jumia Wines In Lagos.csv")
```
